```python
# Making a list of missing value types
missing_values = ["n/a", "na", "--"]
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/dataoptimal/posts/master/data%20cleaning%20with%20python%20and%20pandas/property%20data.csv",
                na_values = missing_values)
```

```python
df
```

```python
#Exception Handling
count=0    #Counter
for row in df['OWN_OCCUPIED']: #MEMBERSHIP OPERATOR
    try: 
        int(row)
        df.loc[count, 'OWN_OCCUPIED']=np.nan
    except ValueError:
        pass
    count= count + 1
```

```python
df
```

```python
count=0
for row in df['NUM_BATH']:
    try: 
        int(row) 
        pass
    except ValueError:
        df.loc[count,'NUM_BATH']=np.nan
    count += 1
```

```python
df
```

```python
df['PID'] = df['PID'].interpolate()
```

```python
df
```

```python
df['ST_NUM'] = df['ST_NUM'].fillna(df.ST_NUM.median())
#df['ST_NUM'] = df['ST_NUM'].interpolate()
```

```python
df
```

```python
df['OWN_OCCUPIED'] = df['OWN_OCCUPIED'].fillna('Y')
```

```python
df
```

```python
#df['NUM_BEDROOMS'] = df['NUM_BEDROOMS'].astype(float).astype(object)
#df['NUM_BEDROOMS'] = df['NUM_BEDROOMS'].astype(object).astype(str)
df['NUM_BEDROOMS'] = df['NUM_BEDROOMS'].fillna(df.NUM_BEDROOMS.median())
```

```python
df['NUM_BATH'] = df['NUM_BATH'].fillna(df.NUM_BATH.median())
```
